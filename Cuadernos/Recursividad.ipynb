# Recursividad

## Ejemplos con sentencias iterativas (ciclos)

### Conteo descendente

Para contar desde un número *n* de uno en uno descendentemente, la técnica iterativa o cíclica repite una misma instrucción utilizando una sentencia como ***while***. En esta manera de resolver el conteo, una variable inicialmente toma un valor, y con cada vuelta o iteración, va cambiando su valor, acercándose hasta su valor final:

```python
variable = 10               # Valor inicial de la variable
while variable >= 0:        # Valor final al que llegara la variable
    print( variable )       # Se muestra el valor de la variable, en cada iteracion
    variable = variable - 1 # La variable incrementa su valor en uno, con cada iteracion
```

### Conteo ascendente

Para contar desde un número *n* de uno en uno descendentemente, la técnica iterativa o cíclica repite una misma instrucción utilizando una sentencia como ***while***. En esta manera de resolver el conteo, una variable inicialmente toma un valor, y con cada vuelta o iteración, va cambiando su valor, acercándose hasta su valor final:

```python
variable = 0                # Valor inicial de la variable
while variable <= 10:       # Valor final al que llegara la variable
    print( variable )       # Se muestra el valor de la variable, en cada iteracion
    variable = variable + 1 # La variable incrementa su valor en uno, con cada iteracion
```

### Multiplicación de dos números

Multiplicar dos números **x** y **n** es similar que sumar **n** veces **x**. Por lo tanto, puede utilizarse una sentencia cíclica que cuente n veces la suma acumulada de **x**:

```python
x = 3
n = 5
suma = 0                    # Toda suma acumulada comienza en cero
for i in range(n):          # n veces
    print(f'suma = {suma} + {x}')
    suma = suma + x         # se acumula x, con cada cilco

print('\n\n')
print(f'suma = {suma}')
```

### Elevar *x* a potencia *n*

Elevar **x** a una potencia **n** es similar que multiplicar **n** veces **x**. Por lo tanto, puede utilizarse una sentencia cíclica que cuente n veces la multiplicación acumulada de **x**:

```python
x = 2
n = 8
potencia = 1                              # Toda multiplicacion acumulada comienza en uno
for i in range(n):                        # n veces
    print(f'potencia = {potencia} * {x}')
    potencia = potencia * x               # se acumula la multiplicacion de x, con cada cilco

print('\n\n')
print(f'potencia = {potencia}')
```

### Suma de numeros entre **n** y **0**

Para calcular la suma acumulada de todos los números enteros entre **n** y **0**, puede utilizarse el ciclo de un contador desde **n** hasta **0** (desde **0** hasta **n**) acumulando la suma de estos valores del contador:


```python
n = 10
suma = 0                            # Toda suma acumulada comienza en cero
for i in range(n, -1, -1):          # desde n hasta 0 (Hay que recordar que Python excluye 0; por eso debe llegar el ciclo  hasta uno antes de cero)
    print(f'suma = {suma} + {i}')
    suma = suma + i                 # se acumula la suma de i, con cada cilco

print('\n\n')
print(f'suma = {suma}')
```

### factorial( n )

Calcular el factorial de **n** es el proceso de multiplicar todos los números entre **n** y **1**. Por lo tanto, puede utilizarse el ciclo de un contador desde **1** hasta **n** (desde **n** hasta **1**) acumulando la multiplicación de estos valores del contador:


```python
n = 5
factorial = 1                               # Toda multiplicacion acumulada comienza en uno
for i in range(1, n+1):                     # desde uno hasta n (Hay que recordar que Python excluye n; por eso debe llegar el ciclo  hasta uno despues de n)
    print(f'factorial = {factorial} * {i}')
    factorial = factorial * i               # se acumula la multiplicacion de i, con cada cilco

print('\n\n')
print(f'{n}! = {factorial}')
```

### fibonacci( n )

Calcular el fibonacci de **n** es el proceso de sumar acumuladamente los valores de **a** y de **b** **n** veces, sustituyendo el valor **a** por el de **b** y el de **b** por el resultado de su suma:

<sub>Nota: esta suma acumulada sólo se hace para valores de **n** mayores a uno; dado que, si n vale lo mismo que alguno de los valores iniciales de **a** o de **b**, su fibonacci sería ese mismo.</sub>

```python
n = 10

a = 0
b = 1
fibonnaci = 0

for i in range( n+1 ):                      # desde cero hasta n (Hay que recordar que Python excluye n; por eso debe llegar el ciclo  hasta uno despues de n)
  if n==0 or n==1:                          # si n no es uno de los valores iniciales de a o de b,
    fibonnaci = n                           # el fibonacci es el mismo valor de n
    break                                   # y no se hacen acumulaciones

  elif i>1:                                 # La suma acumulada debe realizarse a partir de que el contador sea mayor que el inicial de b (o sea, de uno)
    print(f'fibonnaci = {a} + {b}')
    fibonnaci = a+b                         # se acumula la suma de a mas b,
    a = b                                   # se sustituye el valor de a por el de b
    b = fibonnaci                           # se sustituye el valor de b por el de la suma de a con b (llamada aqui fibonacci)

print('\n\n')
print(f'fibonnaci = {fibonnaci}')
```

## Ejemplos con funciones recursivas

### Conteo descendente

En un proceso recursivo, se define una función que invoque a otra idéntica a ella. La primera vez que se invoca a la función recursiva, ésta recibe como argumento un valor; cuando ella invoca a otra, el valor que le envía es menor al suyo. De este modo, cada función manda llamar a su vez a otra con un valor cada vez más cercano al valor final (o de paro) hasta que se alcanza una condición y no se invoca más a ninguna otra función:

```python
def funcion(n):
    print(f'funcion({n}) ', end='')             # Se muestra el valor que recibe ESTA función

    if n>0:                                     # Si el valor actual no es el final,
      print(f'invoca a funcion({n-1})')
      funcion(n-1)                              # se invoca a otra función idéntica, con un valor más cercano al final
    else:
      print(f'ya NO invoca a ninguna funcion')  # Si se ha alcanzao el valor final, la función NO invoca a otra, y termina

n = 10
print(f'Se invoca a funcion({n})\n\n')
funcion( n )                                    # Se invoca a una función enviando el valor de 10
```

### Conteo ascendente

En un proceso recursivo, se define una función que invoque a otra idéntica a ella. La primera vez que se invoca a la función recursiva, ésta recibe como argumento un valor; cuando ella invoca a otra, el valor que le envía es mayor al suyo. De este modo, cada función manda llamar a otra con un valor cada vez más cercano al valor final (o de paro) hasta que se alcanza y no se invoca más a ninguna otra función:

```python
def funcion(n):
    print(f'funcion({n}) ', end='')             # Se muestra el valor que recibe ESTA función

    if n<10:                                     # Si el valor actual no es el final,
      print(f'invoca a funcion({n+1})')
      funcion(n+1)                              # se invoca a otra función idéntica, con un valor más cercano al final
    else:
      print(f'ya NO invoca a ninguna funcion')  # Si se ha alcanzao el valor final, la función NO invoca a otra, y termina

n = 0
print(f'Se invoca a funcion({n})\n\n')
funcion( n )                                    # Se invoca a una función enviando el valor de 10
```

### Multiplicación de dos números

Multiplicar dos números **x** y **n** es similar que sumar **n** veces **x**. Por lo tanto, puede definirse una funcion que retorne la suma del valor que reciba más el de **x**, hasta que se haya invocado **n** funciones:

```python
def funcion(x, n):
  print(f'La función con x={x}, n={n} ', end='')
  if n>0:                                           # Si el valor recibido por una funcion NO ha llegado a cero
    print(f'retorna {x} + funcion({x}, {n-1})')
    return x + funcion(x, n-1)                      # se invoca a otra funcion con un valor cada vez mas cercano, y
                                                    # se retorna la suma de x con el retornado por la funcion invocada

  else:                                             # Si el valor recibido por una funcion HA llegado a cero
    print(f'retorna {n}')
    return n                                        # Se retorna al mismo cero para que la primera suma retornada sea x mas cero


x = 3
n = 5
multiplicacion = funcion(x, n)                      # Se invoca a una función enviando los valores de x y n
print('\n\n')
print(f'{x} * {n} = {multiplicacion}')
```

### Elevar *x* a potencia *n*

Elevar **x** a una potencia **n** es similar que multiplicar **n** veces **x**. Por lo tanto, puede definirse una funcion que retorne la multiplicación del valor que reciba por el de **x**, hasta que se haya invocado **n** funciones:

```python
def funcion(x, n):
  print(f'La función con x={x}, n={n} ', end='')
  if n>0:                                           # Si el valor recibido por una funcion NO ha llegado a cero
    print(f'retorna {x} * funcion({x}, {n-1})')
    return x * funcion(x, n-1)                      # se invoca a otra funcion con un valor cada vez mas cercano, y
                                                    # se retorna la multiplicacion de x con el retornado por la funcion invocada

  else:                                             # Si el valor recibido por una funcion HA llegado a cero
    print(f'retorna 1')
    return 1                                        # Se retorna x a la cero, o sea que se retorna uno


x = 2
n = 8
potencia = funcion(x, n)                      # Se invoca a una función enviando los valores de x y n
print('\n\n')
print(f'{x} ^ {n} = {potencia}')
```

### Suma de numeros entre **n** y **0**

Para calcular la suma de los números entre **n** y **0** puede definirse una funcion que retorne la suma del valor que reciba más el que a su vez le retorne otra función invocada con su valor menos uno. El proceso debe continuar hasta que la última función invocada reciba cero*

```python
def funcion( n ):
  print(f'La función con n={n}\t', end='')
  if n>0:                                   # Si el valor recibido por una funcion NO ha llegado a cero
    print(f'retorna {n}\t+ funcion({n-1})')
    return n + funcion( n-1 )               # se invoca a otra funcion con un valor cada vez menor, y
                                            # se retorna la suma de n con el retornado por la funcion invocada

  else:                                     # Si el valor recibido por una funcion HA llegado a cero
    print(f'retorna {n}')
    return n                                # Se retorna este mismo;


n = 10
suma = funcion(n)                      # Se invoca a una función enviando el valor de n
print('\n\n')
print(f'suma = {suma}')
```

### factorial( n )

Calcular el factorial de **n** es el proceso de multiplicar todos los números entre **n** y **1**. Por lo tanto, puede definirse una funcion que retorne la multiplicación del valor que reciba por el que a su vez le retorne otra función invocada con su valor menos uno. El proceso debe continuar hasta que la última función invocada reciba cero*

<sub>* Recuerde que por definición, el factorial de **cero** es **uno**</sub>


```python
def funcion( n ):
  print(f'La función con n={n} ', end='')
  if n>0:                                   # Si el valor recibido por una funcion NO ha llegado a cero
    print(f'retorna {n} * funcion({n-1})')
    return n * funcion( n-1 )               # se invoca a otra funcion con un valor cada vez menor, y
                                            # se retorna la multiplicacion de n con el retornado por la funcion invocada

  else:                                     # Si el valor recibido por una funcion HA llegado a cero
    print(f'retorna 1')
    return 1                                # Se retorna uno; porque, por definicion, el factorial de cero es uno


n = 5
factorial = funcion(n)                      # Se invoca a una función enviando el valor de n
print('\n\n')
print(f'{n}! = {factorial}')
```

### fibonacci( n )

Calcular el fibonacci de **n** es el proceso de sumar lo que retorne otra función que calcule el fibonacci de **n-1** mas lo que retorne otra que calcule el fibonacci de **n-2**. Este proceso para cuando se invoca una función para el fibonnacci de **uno** de **cero**, cuyo valor es el mismo.

<sub>Nota: las líneas con print fueron comentadas en el código siguiente, porque el cálculo recursivo de fibonacci invoca exponencialmente funciones</sub>

```python
def funcion(n):
  # print(f'La función con n={n} ', end='')
  if n==0 or n==1:                                        # Si el valor recibido por una funcion es cero o uno,
    # print(f'retorna {n}')
    return n                                              # se retorna el mismo
  else:
    # print(f'retorna funcion({n-1}) + funcion({n-2})')
    return funcion(n-1) + funcion(n-2)                    # se invoca a otras dos funciones con un valor cada vez menor, y
                                                          # se retorna la suma

n = 10
fibonacci = funcion( n )

print('\n\n')
print(f'fibonnaci( {n} ) = {fibonnaci}')
```
